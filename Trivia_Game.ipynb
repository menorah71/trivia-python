```python
# Cargar las preguntas desde el archivo JSON
def cargar_preguntas(archivo):
    with open(archivo, "r", encoding="utf-8") as f:
        preguntas = json.load(f)
    return preguntas

# Probamos que cargue bien
preguntas = cargar_preguntas("preguntas.json")

print(f"✅ {len(preguntas)} preguntas cargadas\n")
print("Vista previa de la primera pregunta:")
print(f"  Pregunta:  {preguntas[0]['pregunta']}")
print(f"  Opciones:  {preguntas[0]['opciones']}")
print(f"  Respuesta: {preguntas[0]['respuesta']}")
```

```python
def hacer_pregunta(pregunta_data, numero, total):
    print(f"\n{'='*50}")
    print(f"Pregunta {numero}/{total}")
    print(f"{'='*50}")
    print(f"\n{pregunta_data['pregunta']}\n")
    
    # Mostrar las opciones
    for opcion in pregunta_data['opciones']:
        print(f"  {opcion}")
    
    # Pedir respuesta y validarla
    while True:
        try:
            respuesta = input("\nTu respuesta (a/b/c/d): ").lower().strip()
            
            if respuesta not in ['a', 'b', 'c', 'd']:
                print("⚠️ Solo puedes escribir a, b, c o d")
                continue
                
            # Verificar si es correcta
            if respuesta == pregunta_data['respuesta']:
                print("✅ ¡Correcto! +10 puntos")
                print(f"💡 {pregunta_data['explicacion']}")
                return 10
            else:
                correcta = pregunta_data['respuesta']
                print(f"❌ Incorrecto. La respuesta era: {correcta})")
                print(f"💡 {pregunta_data['explicacion']}")
                return 0
                
        except Exception as e:
            print(f"Error inesperado: {e}")
            continue

print("✅ Función lista!")
```

```python
def jugar():
    # Bienvenida
    print("="*50)
    print("   BIENVENIDO AL JUEGO DE TRIVIA PYTHON!")
    print("="*50)
    
    nombre = input("\nEscribe tu nombre: ").strip()
    print(f"\nHola {nombre}! Vamos a ver cuanto sabes...\n")
    
    # Cargar y mezclar preguntas
    todas = cargar_preguntas("preguntas.json")
    random.shuffle(todas)          # mezcla el orden cada partida
    seleccionadas = todas[:5]      # tomamos solo 5 preguntas
    
    # Jugar
    puntaje = 0
    total = len(seleccionadas)
    
    for i, pregunta_data in enumerate(seleccionadas, start=1):
        puntos = hacer_pregunta(pregunta_data, i, total)
        puntaje += puntos
    
    # Resultado final
    print("\n" + "="*50)
    print(f"   JUEGO TERMINADO, {nombre}!")
    print("="*50)
    print(f"\nPuntaje final: {puntaje}/{total * 10}")
    
    # Nivel según puntaje
    if puntaje == 50:
        nivel = "🏆 Maestro Python!"
    elif puntaje >= 40:
        nivel = "🥇 Experto!"
    elif puntaje >= 30:
        nivel = "🥈 Avanzado!"
    elif puntaje >= 20:
        nivel = "🥉 En progreso!"
    else:
        nivel = "📚 Sigue estudiando!"
    
    print(f"Nivel: {nivel}\n")

# Iniciamos el juego!
jugar()
```
